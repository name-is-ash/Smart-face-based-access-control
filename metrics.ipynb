```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python

# Assume these metrics are computed as in the main_video.py code
accuracy = 1.00
precision = 1.00
recall = 1.00
f1 = 1.00
cm = np.array([[13, 0], [0, 10]])  # Example confusion matrix for Known and Unknown faces

# Known and unknown faces metrics
known_faces_accuracy = 1.00
unknown_faces_detection_rate = 1.00
```

```python

# 1. Plot Confusion Matrix as Heatmap
def plot_confusion_matrix(cm):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Manoj", "Prasanna"], 
                yticklabels=["Manoj", "Prasanna"])
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.show()
```

```python

# 2. Bar Plot for Accuracy, Precision, Recall, F1-Score
def plot_evaluation_metrics(accuracy, precision, recall, f1):
    metrics = {"Accuracy": accuracy, "Precision": precision, "Recall": recall, "F1-Score": f1}
    plt.figure(figsize=(8, 5))
    plt.bar(metrics.keys(), metrics.values(), color=['blue', 'green', 'orange', 'red'])
    plt.ylim(0, 1.1)
    plt.title("Evaluation Metrics")
    plt.ylabel("Score")
    plt.show()
```

```python

# 3. Pie Chart for Class Distribution
def plot_class_distribution(cm):
    class_sums = np.sum(cm, axis=1)
    labels = ["Manoj", "Prasanna"]
    plt.figure(figsize=(6, 6))
    plt.pie(class_sums, labels=labels, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
    plt.title("Class Distribution of Predictions")
    plt.show()
```

```python

# 4. Line Plot for Time Duration of Recognized Faces
# Example data for durations
names = ["Manoj", "Prasanna"]
durations = [10, 13]  # Example durations

def plot_duration_per_face(names, durations):
    plt.figure(figsize=(8, 6))
    plt.plot(names, durations, marker='o', linestyle='-', color='b')
    plt.title("Recognition Duration per Face")
    plt.xlabel("Face Name")
    plt.ylabel("Duration (seconds)")
    plt.show()
```

```python

# Call the plotting functions to visualize
plot_confusion_matrix(cm)
plot_evaluation_metrics(accuracy, precision, recall, f1)
plot_class_distribution(cm)
plot_duration_per_face(names, durations)
```
